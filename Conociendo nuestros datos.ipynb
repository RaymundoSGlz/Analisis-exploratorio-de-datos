````python
cdf_flipper_length_mm = empiricaldist.Cdf.from_seq(
    processed_penguins_df.flipper_length_mm,
    normalize=True
)
````

````python
cdf_flipper_length_mm.plot();
````

````python
#gr√°fica de Cdf
cdf_flipper_length_mm.plot()

#calcular la probabilidad de forma manual
q = 190
p = cdf_flipper_length_mm.forward(q)

#trazo de linea vertical en la gr√°fica
plt.vlines(
    x=q,
    ymin=0,
    ymax=p,
    color='black',
    linestyle='dashed'
)

#trazo de linea horizontal en la gr√°fica
plt.hlines(
    y=p,
    xmin=pmf_flipper_length.qs[0],
    xmax=q,
    color='black',
    linestyle='dashed'
)

plt.plot(q, p, 'ro');
````

````python
#calculo de un valor en funci√≥n de la probabilidad dada
cdf_flipper_length_mm.step()

#especificar probabilidades
p_1 = 0.25
p_2 = 0.75

#IQR
ps = (0.25, 0.75)

#obtener el valor dada la probabilidad
qs = cdf_flipper_length_mm.inverse(ps)

plt.vlines(
    x=qs,
    ymin=0,
    ymax=ps,
    color='black',
    linestyle='dashed'
)

plt.hlines(
    y=ps,
    xmin=pmf_flipper_length.qs[0],
    xmax=qs,
    color='black',
    linestyle='dashed'
)

plt.scatter(
    x=qs,
    y=ps,
    color='red',
    zorder=2
);
````

# Ley de los Grandes N√∫meros y Teorema del L√≠mite Central

## Ley de los Grandes N√∫meros

````python
#calculo de la probabilidad para cada cara del dado
dice = empiricaldist.Pmf.from_seq([1,2,3,4,5,6])
dice.bar()
````

````python
for sample_size in (1e2, 1e3,1e4):  #100 1000 10000
    sample_size = int(sample_size)
    values = dice.sample(sample_size) # obtener valores seg√∫n tama√±o de la muestra
    sample_pmf = empiricaldist.Pmf.from_seq(values) #calculo de la probabilidad para cada valor
	
	#gr√°ficas para cada tama√±o de muestra
    plt.figure(figsize=(5, 5))
    sample_pmf.bar()
    plt.axhline(y=1/6, color='red', linestyle='dashed')
    plt.ylim([0,0.50])
    plt.title(f'Sample size: {sample_size}')
````

# Limitaciones del an√°lisis de regresi√≥n simple

## La regresi√≥n lineal simple no es sim√©trica

````python
x = processed_penguins_df.bill_length_mm
y = processed_penguins_df.bill_depth_mm

res_x_y = scipy.stats.linregress(x=x, y=y)
res_y_x = scipy.stats.linregress(x=y, y=x)

print(res_x_y, res_y_x, sep='\n')
````

````python
#Regresi√≥n lineal de x en funcion de y
sns.scatterplot(
    x=x,
    y=y
)

fx_1 = np.array([x.min(), x.max()])
fy_1 = res_x_y.intercept + res_x_y.slope*fx_1

plt.plot(fx_1, fy_1);
````

````python
#Regresi√≥n lineal de y en funcion de x
sns.scatterplot(
    x=y,
    y=x
)

fx_2 = np.array([y.min(), y.max()])
fy_2 = res_y_x.intercept + res_y_x.slope*fx_2

plt.plot(fx_2, fy_2);
````

## Olvid√© mi b√°scula para pesar a los ping√ºinos, ¬øcu√°l ser√≠a la mejor forma de capturar ese dato?

### Creando modelos

#### Modelo 1

````python
# Peso de los penguins vs longitud del pico
model_1 = (
    smf.ols(
        formula='body_mass_g ~ bill_length_mm',
        data=processed_penguins_df
    )
    .fit()
)
model_1.summary() # descripci√≥n del modelo
````

#### Modelo 2

````python
# Peso del pinguen vs longitud del pico, ancho del pico
model_2 = (
    smf.ols(
        formula='body_mass_g ~ bill_length_mm + bill_depth_mm',
        data=processed_penguins_df
    )
    .fit()
)
model_2.summary() # descripci√≥n del modelo
````

#### Modelo 3

````python
# Peso de los penguins vs longitud del pico, ancho del pico, longitud de las alas
model_3 = (
    smf.ols(
        formula='body_mass_g ~ bill_length_mm + bill_depth_mm + flipper_length_mm',
        data=processed_penguins_df
    )
    .fit()
)
model_3.summary() # descripci√≥n del modelo
````

#### Modelo 4

````python
# Peso de los penguins vs longitud del pico, ancho del pico, longitud de las alas, sexo del pinguen donde sexo es de tipo categ√≥rica
model_4 = (
    smf.ols(
        formula='body_mass_g ~ bill_length_mm + bill_depth_mm + flipper_length_mm + C(sex)',
        data=processed_penguins_df
    )
    .fit()
)

model_4.summary()
````

#### Modelo 5

````python
# Peso del pinguen vs longitud de las alas, sexo
model_5 = (
    smf.ols(
        formula='body_mass_g ~ flipper_length_mm + C(sex)',
        data=processed_penguins_df
    )
    .fit()
)

model_5.summary()
````
